```python
import math

# input values
wind_speed = float(input("Enter wind speed (m/s): "))
blade_radius = float(input("Enter blade radius (m): "))

air_density = 1.225

# swept area
area = math.pi * blade_radius ** 2

# power calculation
power = 0.5 * air_density * area * wind_speed ** 3

print("Power Output:", power, "Watts")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

blade_radius = 40
air_density = 1.225

area = math.pi * blade_radius ** 2

wind_speed = np.arange(1, 25)

power = 0.5 * air_density * area * wind_speed ** 3

plt.plot(wind_speed, power)

plt.xlabel("Wind Speed (m/s)")
plt.ylabel("Power Output (Watts)")
plt.title("Wind Turbine Power Output")

plt.show()
```

```python
import pandas as pd

data = pd.DataFrame({
    "Wind Speed": wind_speed,
    "Power": power
})

data.to_csv("wind_power_results.csv", index=False)
```
